```python
news_article=[]
for x in range(1, 6):

    html = browser.html
    soup = BeautifulSoup(html, 'lxml')
    
    head_title=soup.find_all('div', class_='content_title')

    for title in head_title:
        data_file=title.text.strip()
        news_article.append(data_file)
        print(data_file)

    #browser.click_link_by_partial_text('Next')
```
